# IV. Examine the connectedness of KEGG subgraphs, systematically

```python
graph_names = ["inbiomap_experimentally_graph", 
               "inbiomap_full_graph", 
               "iref13_graph", 
               "iref14_graph", 
               "iref13_iref14_merged_graph"]

graphs = [inbiomap_experimentally_graph,
          inbiomap_full_graph,
          iref13_graph,
          iref14_graph,
          iref13_iref14_merged_graph]
```

```python
gene_lists = list(map(list, pathways.values()))
```

```python
def connectedness(subG): 
    num_subG_nodes = subG.number_of_nodes()
    num_nodes_in_complete_graph_with_size_subG = num_subG_nodes * (num_subG_nodes - 1) / 2
    return (subG.number_of_edges() / (num_nodes_in_complete_graph_with_size_subG + 1))
    
def connectednesses(G): return list(map(connectedness, [G.subgraph(genes) for genes in gene_lists]))
```

```python
graph_connectednesses = pd.DataFrame(dict(zip(graph_names, [connectednesses(graph) for graph in graphs])), index=pathways.keys())
graph_connectednesses.head()
```

```python
graph_connectednesses.plot.hist(alpha=0.3, figsize=(17,8), bins=50)
```

```python
graph_connectednesses.plot.box(figsize=(17,8))
```

### This figure suggests many of the graphs contain the same information about each of the pathways 

```python
graph_connectednesses.plot.area(stacked=False, figsize=(17,8))
```

### This figure shows there's still variability but there *does* exist a moderate trend

```python
graph_connectednesses.sort_values('inbiomap_full_graph', ascending=False).plot.area(stacked=False, figsize=(17,8))
```

# V. Examine the number of components in KEGG pathways, systematically

```python
def disconnectedness(subG): return nx.number_connected_components(subG) / (subG.number_of_nodes() + 1)
    
def disconnectednesses(G): return list(map(disconnectedness, [G.subgraph(genes) for genes in gene_lists]))
```

```python
graph_connectednesses = pd.DataFrame(dict(zip(graph_names, [disconnectednesses(graph) for graph in graphs])), index=pathways.keys())
graph_connectednesses.plot.hist(alpha=0.3, figsize=(17,8), bins=50)
```
